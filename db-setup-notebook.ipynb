## Get historical event information from The Blue Alliance
TBA API Documentation: https://www.thebluealliance.com/apidocs/v3

```python
from datetime import date
from tba_handler import get_tba_response

eventsList = []
allEvents = []

for year in range(2000, 2027):
    eventResponse = get_tba_response(f"/events/{year}/simple")
    responselist = eventResponse.json()
    for event in responselist:
        allEvents.append(event)
        if event['state_prov'] == 'MI':
            eventsList.append(event)

# ## Some Notes
# "Event Type" seems to map to one of a few options:
# {
#     0 : regional_event
#     1 : district_event
#     2 : district_championship
#     3 : championship_division
#     4 : championship_event #With some exceptions, this has denoted the Einstein Field since 2001
# }

# ## Adding Data I Care About
# One additional thing I'm interested in is whether the event ends on a Saturday or Sunday, but I'll need to figure that out programmatically
```

```python
for i in range(0, len(eventsList)):
    endDateStr = eventsList[i]['end_date']
    endDateObj = date.fromisoformat(endDateStr)
    eventsList[i]['end_DoW'] = endDateObj.strftime("%a")
```

```python
## Events to Database
import dataset
from dotenv import load_dotenv
import os

load_dotenv()

dbpath = os.getenv("dbpath")
db = dataset.connect(dbpath)

# Define the database schema
'''
Table events {
  id int [primary key, not null, unique]
  city text
  country text
  district dictrict_object
  end_date numeric
  event_code text
  event_type int
  key text [not null, unique]
  name text
  start_date numeric
  state_prov text
  year int [not null]
  end_DoW text [not null]
}

Table people {
  id int [primary key, unique]
  first_name text [not null]
  last_name text [not null]
  phone numeric
  rookie_season int [default: null]
  ref_role enum
}

Table users {
  id int [primary key, unique]
  person_id int [unique]
  site_role enum
}
'''
```
